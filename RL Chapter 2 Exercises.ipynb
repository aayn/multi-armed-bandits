## Exercise 2.5

Design and conduct an experiment to demonstrate the
difficulties that sample-average methods have for nonstationary problems.
Use a modified version of the 10-armed testbed in which all the q\*(a) start
out equal and then take independent random walks (say by adding a normally
distributed increment with mean zero and standard deviation 0.01 to all the
q\*(a) on each step). Prepare plots like Figure 2.2 for an action-value
method using sample averages, incrementally computed, and another
action-value method using a constant step-size parameter, alpha = 0.1. Use

## Exercise 2.11

Make a figure analogous to Figure 2.6 for the nonstationary case outlined in
Exercise 2.5. Include the constant-step-size eps-greedy algorithm with alpha =
0.1. Use runs of 200,000 steps and, as a performance measure for each algorithm
and parameter setting, use the average reward over the last 100,000 steps.

```python
# Incrementally-computed sample average
p1s = [float(pow(2, i)) for i in range(-7, -1)]
a1s = [alg.EpsilonGreedy(eps=eps) for eps in p1s]

# Constant-step sample average
p2s = [float(pow(2, i)) for i in range(-7, -1)]
a2s = [alg.EpsilonGreedy(eps=eps, alpha=0.1) for eps in p2s]

# With optimistic initialization
p3s = [float(pow(2, i)) for i in range(-2, 3)]
a3s = [alg.EpsilonGreedy(eps=0.1, alpha=0.1, Q1=Q1) for Q1 in p3s]

# UCB
p4s = [float(pow(2, i)) for i in range(-4, 3)]
a4s = [alg.UCB(c=c) for c in p4s]

# Gradient Bandit
p5s = [float(pow(2, i)) for i in range(-5, 2)]
a5s = [alg.GradientBandit(alpha=alpha) for alpha in p5s]

all_pvalues = [p1s, p2s, p3s, p4s, p5s]
all_algs = [a1s, a2s, a3s, a4s, a5s]
```

```python
# Set runs = 100, and steps = 200000 in config.yml
results = []
for al in all_algs:
    inputs = [(a, True) for a in al]
    result = multi_run(*inputs)
    results.append(result)
```

```python
all_vals = []
for result in results:
    vals = [np.average(ar[100001:]) for ar, _ in result]
    all_vals.append(vals)
```

```python
labels = ['Incremental sample average; parameter: eps',
          'Constant-step sample average; parameter: eps',
          'Optimistic initialization; parameter: Q0',
          'UCB; parameter: c',
          'Gradient Bandit; parameter: alpha']
plot.parameter_study(all_vals, all_pvalues, labels)
```
